### 7. Group By 

#### a. How many of body-styles that each car brand / company have? 

```python
df.groupby('company')[['body-style']].count()
```

#### b. Find each company's highest car price

```python
df.groupby('company')[['price']].max()
```

#### c. Find the average mileage per engine-type and body style

```python
df.groupby(['engine-type','body-style'])[['average-mileage']].mean()
```

#### d.  How many records fall into each company and body-style combination? 

```python
df.groupby(['company', 'body-style']).count()
```

### 8. Pivot Tables

#### a. Find the average mileage, horsepower, length and price for each company

```python
df.pivot_table(index = 'company', values=['average-mileage','horsepower','length','price'])
```

#### b. Find the maximum horsepower for each company and engine-type

```python
df.pivot_table(index = 'company', values='horsepower', columns='engine-type', aggfunc=max)
```

#### c. Find the lowest and highest prices for each company and body style

```python
df.pivot_table(index = 'company', values='price', columns='body-style', aggfunc=[max,min])
```
